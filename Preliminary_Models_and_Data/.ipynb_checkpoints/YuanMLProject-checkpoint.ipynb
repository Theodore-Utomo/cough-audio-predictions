```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df_coughvid = pd.read_csv('/content/cough-audio-predictions/tabular_form/coughvid_v3.csv')
df_extracted_features = pd.read_csv('/content/cough-audio-predictions/tabular_form/extracted_features_coughvid_v3.csv')
df_filtered_labels = pd.read_csv('/content/cough-audio-predictions/tabular_form/filtered_expert_labels_coughvid_v3.csv')
```

```python
from sklearn.preprocessing import StandardScaler

df_clean = df_extracted_features.dropna()

X = df_clean.drop(columns=['status'])
y = df_clean['status']

if 'gender' in X.columns:
    X['gender'] = X['gender'].map({'male': 0, 'female': 1})

X = pd.get_dummies(X, drop_first=True)

# Replacing infinite and nan values
X = X.replace([np.inf, -np.inf], np.nan).dropna()

# IMPORTANT: Re-align y to the cleaned X
y = y.loc[X.index]

# One-hot encode the target variable
y_encoded = pd.get_dummies(y, prefix='status')
print("One-hot encoded target:\n", y_encoded.head())

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)
print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
# Normalize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Check for NaNs after scaling
assert not np.isnan(X_train_scaled).any(), "Found NaNs in X_train_scaled"
assert not np.isnan(X_test_scaled).any(), "Found NaNs in X_test_scaled"


# Convert DataFrames to numpy arrays, then to torch tensors
X_train_np = X_train_scaled.astype(np.float32)
X_test_np = X_test_scaled.astype(np.float32)
y_train_np = y_train.to_numpy().astype(np.float32)
y_test_np = y_test.to_numpy().astype(np.float32)

X_train_tensor = torch.tensor(X_train_np, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test_np, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train_np, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test_np, dtype=torch.float32)

# Reshape X to add a channel dimension (for 1D CNN input)
X_train_tensor = X_train_tensor.unsqueeze(1)
X_test_tensor = X_test_tensor.unsqueeze(1)


# Create TensorDataset and DataLoader objects
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)
batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Model Definition 

class CNNClassifier(nn.Module):
    def __init__(self, num_features, num_classes):
        super(CNNClassifier, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=16, kernel_size=3, padding=1)
        self.relu1 = nn.ReLU()
        self.conv2 = nn.Conv1d(in_channels=16, out_channels=32, kernel_size=3, padding=1)
        self.relu2 = nn.ReLU()
        self.pool = nn.AdaptiveMaxPool1d(1)
        self.fc = nn.Linear(32, num_classes)

    def forward(self, x):
        x = self.conv1(x)
        x = self.relu1(x)
        x = self.conv2(x)
        x = self.relu2(x)
        x = self.pool(x)
        x = x.squeeze(2)
        x = self.fc(x)
        return x

num_features = X_train_tensor.shape[2]
num_classes = y_train_tensor.shape[1]
model = CNNClassifier(num_features, num_classes)
print(model)

# Loss Function and Optimizer 

y_train_indices = torch.argmax(y_train_tensor, dim=1)
y_test_indices = torch.argmax(y_test_tensor, dim=1)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)  # Lower learning rate

# Training Loop 

num_epochs = 50
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, targets in train_loader:
        targets_indices = torch.argmax(targets, dim=1)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets_indices)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * inputs.size(0)
    epoch_loss = running_loss / len(train_loader.dataset)
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}")


model.eval()
correct = 0
total = 0
with torch.no_grad():
    for inputs, targets in test_loader:
        targets_indices = torch.argmax(targets, dim=1)
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        total += targets.size(0)
        correct += (predicted == targets_indices).sum().item()

print("Test Accuracy: {:.2f}%".format(100 * correct / total))
```
